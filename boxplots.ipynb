```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.plot(kind='box')
```

```python
my_df.plot(kind='box', figsize=(10, 5))
```

```python
my_df.plot(kind='box', cmap='jet')
```

```python
my_df.plot(kind='box', legend=True)
```

```python
my_df.plot(kind='box', title='Box Plot')
```

```python
my_df.plot.box()
```

```python
my_df.plot(kind='box'grid=True)
```
