```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
my_df.plot(kind='hexbin', x='Mon', y='Tues')
```

```python
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10)
```

```python
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10, cmap='Wistia')
```

```python
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10)
```

```python
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10, title='Hex Plot')
```

```python
my_df.plot.hexbin(x='Mon', y='Tues', C='Wed', gridsize=10, title='Hex Plot', grid=True)
```
