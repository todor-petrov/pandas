```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

my_df = pd.DataFrame(randn(500,4), columns=['Mon', 'Tues', 'Wed', 'Thurs'])
my_df
```

```python
my_df.plot(kind='kde')
```

```python
my_df.plot.kde()
```

```python
my_df.plot.density()
```

```python
my_df.plot(kind='density')
```

```python
my_df['Mon'].plot.kde(alpha=0.4, lw=4)
```

```python
my_df['Mon'].plot.kde(alpha=0.4, lw=4, grid=True)
```

```python
my_df['Mon'].plot.kde(alpha=0.4, lw=4, title='')
```
