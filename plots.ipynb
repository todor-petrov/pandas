```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
my_df = pd.DataFrame(randn(25, 4), columns=['Mon', 'Tues', 'Wed', 'Thur'])
my_df
```

```python
my_df.abs().plot(kind='area')
```

```python
my_df.abs().plot(kind='area', alpha=0.5)
```

```python
my_df.abs().plot(kind='area', alpha=0.5, title='Awesome Area Plot')
```

```python
my_df.abs().plot(kind='area', alpha=0.5, title='Awesome Area Plot', legend=False)
```

```python
my_df['Mon'].abs().plot(kind='area', alpha=0.5, title='Awesome Area Plot', legend=False)
```

```python
my_df[['Mon', 'Thur']].abs().plot(kind='area', alpha=0.5, title='Awesome Area Plot', legend=False)
```

```python
my_df[['Mon', 'Thur']].abs().plot(kind='area', alpha=0.5, title='Awesome Area Plot', legend=False, table=True)
```
