```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
my_df = pd.DataFrame(randn(25, 4), columns=['Mon', 'Tues', 'Wed', 'Thur'])
my_df
```

```python
my_df.plot(kind='line')
```

```python
my_df.abs().plot(kind='line')
```

```python
my_df.abs().plot(kind='line', lw=5)
```

```python
my_df['Mon'].abs().plot(kind='line')
```

```python
my_df[['Mon', 'Wed']].abs().plot(kind='line')
```

```python
my_df[['Mon', 'Wed']].abs().plot(kind='line', figsize=(10, 5))
```

```python
my_df[['Mon', 'Wed']].abs().plot(kind='line', title='Line Chart')
```

```python
my_df[['Mon', 'Wed']].abs().plot(kind='line', alpha=0.3)
```

```python
my_df[['Mon', 'Wed']].abs().plot(kind='line', legend=False)
```
